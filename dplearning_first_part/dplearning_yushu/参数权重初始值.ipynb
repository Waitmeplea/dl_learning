```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x=np.random.randn(1000,100)
node_num=100
hidden_layer_size=5
result={}
```

```python
a=x
for i in range(hidden_layer_size):
    node_num=100
    w=np.random.randn(node_num,node_num)*0.1
    z=np.dot(a,w)
    a=sigmoid(z)
    result[i]=a
```

```python
for i ,a in result.items():
    plt.subplot(1,len(result),i+1)
    plt.hist(a.flatten(),30,range=(0,1))
```
