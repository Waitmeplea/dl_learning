```python
from pathlib import Path
import sys
import torch
from modelscope import AutoModelForCausalLM, AutoTokenizer,AutoModel
from pathlib import Path
from modelscope import AutoTokenizer, AutoModelForMaskedLM
model_id='iic/nlp_gte_sentence-embedding_chinese-base'
tokenizer=AutoTokenizer.from_pretrained(model_id)
model = AutoModel.from_pretrained(model_id)
```

```python
#1、加载文件
def txt_2db():

    #0 遍历所有文件
    txt_path_list=list(Path('rag_document').glob('*txt'))
    text_list=[]
    #1、找到和问题相关的文件 知识库有限 用文件去找问题，当知识库很大则需要倒排索引 用问题查文件
    for path in txt_path_list:
        text_list.append(path.read_text(encoding='utf-8'))


    #2、相关文件内容读取出来
    #3、添加到context中
    return text_list
#2、进行向量嵌入
#3、存入向量数据库
```

```python
def embed(sentences):
    inputs = tokenizer(sentences, return_tensors='pt', padding=True, truncation=True)

    with torch.no_grad():
        result=model(**inputs)
    return result['pooler_output']
```
